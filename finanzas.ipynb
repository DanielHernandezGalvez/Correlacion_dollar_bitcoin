```python
import yfinance as yf

# Crear objeto Ticker para Tesla
tesla = yf.Ticker("TSLA")

# Extraer datos históricos
tesla_data = tesla.history(period="max")

# Resetear índice y mostrar primeras filas
tesla_data.reset_index(inplace=True)
print(tesla_data.head())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
headers = {'User-Agent': 'Mozilla/5.0'}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

# Encontrar la tabla de ingresos trimestrales
table = soup.find_all('table')[1]  # Segunda tabla contiene los datos trimestrales
tesla_revenue = pd.read_html(str(table))[0]
tesla_revenue.columns = ['Date', 'Revenue']

# Limpiar datos
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]
print(tesla_revenue.tail())
```

```python
import yfinance as yf

# Crear objeto Ticker para GameStop
gme = yf.Ticker("GME")

# Extraer datos históricos
gme_data = gme.history(period="max")

# Resetear índice y mostrar primeras filas
gme_data.reset_index(inplace=True)
print(gme_data.head())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
headers = {'User-Agent': 'Mozilla/5.0'}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

# Encontrar la tabla de ingresos trimestrales
table = soup.find_all('table')[1]
gme_revenue = pd.read_html(str(table))[0]
gme_revenue.columns = ['Date', 'Revenue']

# Limpiar datos
gme_revenue = gme_revenue[gme_revenue['Revenue'] != ""]
print(gme_revenue.tail())
```

```python
import matplotlib.pyplot as plt

# Gráfico de precios de Tesla
plt.figure(figsize=(14, 7))
plt.plot(tesla_data['Date'], tesla_data['Close'])
plt.title('Historial de Precios de Tesla')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre ($)')
plt.grid()
plt.show()

# Gráfico de ingresos de Tesla
tesla_revenue['Revenue'] = tesla_revenue['Revenue'].str.replace(',', '').str.replace('$', '').astype(float)
plt.figure(figsize=(14, 7))
plt.bar(tesla_revenue['Date'], tesla_revenue['Revenue'])
plt.title('Ingresos Trimestrales de Tesla')
plt.xlabel('Trimestre')
plt.ylabel('Ingresos ($ millones)')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Gráfico de precios de GameStop
plt.figure(figsize=(14, 7))
plt.plot(gme_data['Date'], gme_data['Close'], color='red')
plt.title('Historial de Precios de GameStop')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre ($)')
plt.grid()
plt.show()

# Gráfico de ingresos de GameStop
gme_revenue['Revenue'] = gme_revenue['Revenue'].str.replace(',', '').str.replace('$', '').astype(float)
plt.figure(figsize=(14, 7))
plt.bar(gme_revenue['Date'], gme_revenue['Revenue'], color='green')
plt.title('Ingresos Trimestrales de GameStop')
plt.xlabel('Trimestre')
plt.ylabel('Ingresos ($ millones)')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# Código para guardar todos los DataFrames en un archivo Excel
with pd.ExcelWriter('stock_analysis_results.xlsx') as writer:
    tesla_data.to_excel(writer, sheet_name='Tesla_Stock')
    tesla_revenue.to_excel(writer, sheet_name='Tesla_Revenue')
    gme_data.to_excel(writer, sheet_name='GME_Stock')
```
